```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pd_symm(m=0,d1=2,d2=4):
    global L,D,S,Q,E,e1,e2
    L=np.array([[1,0],
                [m,1]])
    D=np.array([[d1,0],
                [0,d2]])
    S=L@D@L.T
    (e1,e2),Qinv=np.linalg.eig(S)
    Q=Qinv.T
    E=np.eye(2)@[e1,e2]
```

```python
pd_symm(m=0.5,d1=2,d2=4)
print(f'''
L=
{L}
D=
{D}
S=
{S}
Q=
{Q}
E=
{E}
''')
```

```python
mm_s=np.linspace(-2,2,100)
ee1=np.empty(100,dtype=float)
ee2=np.empty(100,dtype=float)
d1=2
d2=4
for i,mm in enumerate(mm_s):
    pd_symm(m=mm,d1=d1,d2=d2)
    ee1[i]=e1
    ee2[i]=e2
plt.plot(mm_s,ee1)
plt.plot(mm_s,ee2)
```

```python
%matplotlib notebook

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(ee1,ee2,mm_s)
# plt.scatter(d1,d2)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
